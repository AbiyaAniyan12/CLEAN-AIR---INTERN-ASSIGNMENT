#### Time-Series Visualisation

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset 
df = pd.read_csv(r"C:\Users\Abiya\OneDrive\Desktop\Internship\SWM DATA.csv")
df.head(7)
```

```python
df.shape
df.info()
df.describe()
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
# Handle missing values in numerical columns by filling with the mean
numeric_columns = df.select_dtypes(include=['float64']).columns
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.fillna(x.mean()))
```

```python
df.head(7)
```

```python
df.isnull().sum()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.title('Municipal Solid Waste Trends', fontsize=14)
# Show plot
plt.grid(True)  # Optional: Add a grid
plt.show()
```

```python
# Plotting the time series
plt.figure(figsize=(10, 6,))
plt.plot(df['YEAR'], df['Treated'], marker='o', label='Treated (%)', color='green')
plt.plot(df['YEAR'], df['Land filled'], marker='o', label='Landfilled (%)', color='blue')
plt.plot(df['YEAR'], df['Unaccounted Percentage'], marker='o', label='Unaccounted (%)', color='red')
# Add title, x-axis and y-axis labels
plt.title('MSW Management Trends (Treated, Landfilled, Unaccounted)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
```

```python
# Adding annotations
plt.title('Municipal Solid Waste Trends (2015–2022)', fontsize=14)
plt.xlabel('YEAR', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.axvline(x=2016, color='orange', linestyle='--', label='Swachh Bharat Launch')
plt.legend()
plt.grid(True)
plt.show()
```
